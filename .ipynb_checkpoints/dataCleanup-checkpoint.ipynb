```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

## Data cleanup

```python
import pandas as pd

df = pd.read_csv('obesity.csv')

df.head()
```

```python
#data cleanup - renamed column with better naming convention
df.rename(columns = {'Gender': 'gender'}, inplace = True)
df.rename(columns = {'Age': 'age'}, inplace = True)
df.rename(columns = {'Height': 'height'}, inplace = True)
df.rename(columns = {'Weight': 'weight'}, inplace = True)
df.rename(columns = {'family_history_with_overweight': 'famHistOverweight'}, inplace = True)
df.rename(columns = {'FAVC': 'highCalFood'}, inplace = True)
df.rename(columns = {'FCVC': 'veggies'}, inplace = True)
df.rename(columns = {'NCP': 'numMealsPerDay'}, inplace = True)
df.rename(columns = {'CAEC': 'foodBetweenMeals'}, inplace = True)
df.rename(columns = {'SMOKE': 'smoke'}, inplace = True)
df.rename(columns = {'CH2O': 'h20PerDay'}, inplace = True)
df.rename(columns = {'SCC': 'trackCals'}, inplace = True)
df.rename(columns = {'FAF': 'physActiFreq'}, inplace = True)
df.rename(columns = {'TUE': 'techFreq'}, inplace = True)
df.rename(columns = {'CALC': 'alcFreq'}, inplace = True)
df.rename(columns = {'MTRANS': 'transp'}, inplace = True)
df.rename(columns = {'NObeyesdad': 'obesityLvl'}, inplace = True)

#display first 25 values
df.head(25)
```

```python
#Data cleanup - replaced values in gender column for easier analyses
df['gender'] = df['gender'].replace('Female','F')
df['gender'] = df['gender'].replace('Male','M')
df.head(25)
```

```python
#data cleanup - changed string to integer values for analyses
df['famHistOverweight'] = df['famHistOverweight'].replace('yes',1)
df['famHistOverweight'] = df['famHistOverweight'].replace('no',0)
df.head(25)
```

```python
#Data Cleanup - changed string values from yes,no to integers 1,0 for easier analyses
df['highCalFood'] = df['highCalFood'].replace('yes',1)
df['highCalFood'] = df['highCalFood'].replace('no',0)
df['smoke'] = df['smoke'].replace('yes',1)
df['smoke'] = df['smoke'].replace('no',0)
df['trackCals'] = df['trackCals'].replace('yes',1)
df['trackCals'] = df['trackCals'].replace('no',0)

warnings.simplefilter('ignore')
df.head(25)
```

```python
#Data Cleanup - updated obesity level values for more uniformity + easier read
df['obesityLvl'] = df['obesityLvl'].replace('Normal_Weight','NormalWeight')
df['obesityLvl'] = df['obesityLvl'].replace('Insufficient_Weight','InsufficientWeight')
df['obesityLvl'] = df['obesityLvl'].replace('Overweight_Level_I','OverweightLvl1')
df['obesityLvl'] = df['obesityLvl'].replace('Overweight_Level_II','OverweightLvl2')
df['obesityLvl'] = df['obesityLvl'].replace('Obesity_Type_I','ObesityType1')
df['obesityLvl'] = df['obesityLvl'].replace('Obesity_Type_II','ObesityType2')
df['obesityLvl'] = df['obesityLvl'].replace('Obesity_Type_III','ObesityType3')

df.head(25)
```

```python
df2 = pd.read_csv('obesity.csv')

#data cleanup - renamed columns
df2.rename(columns = {'Gender': 'gender'}, inplace = True)
df2.rename(columns = {'Age': 'age'}, inplace = True)
df2.rename(columns = {'Height': 'height'}, inplace = True)
df2.rename(columns = {'Weight': 'weight'}, inplace = True)
df2.rename(columns = {'family_history_with_overweight': 'famHistOverweight'}, inplace = True)
df2.rename(columns = {'FAVC': 'highCalFood'}, inplace = True)
df2.rename(columns = {'FCVC': 'veggies'}, inplace = True)
df2.rename(columns = {'NCP': 'numMealsPerDay'}, inplace = True)
df2.rename(columns = {'CAEC': 'foodBetweenMeals'}, inplace = True)
df2.rename(columns = {'SMOKE': 'smoke'}, inplace = True)
df2.rename(columns = {'CH2O': 'h20PerDay'}, inplace = True)
df2.rename(columns = {'SCC': 'trackCals'}, inplace = True)
df2.rename(columns = {'FAF': 'physActiFreq'}, inplace = True)
df2.rename(columns = {'TUE': 'techFreq'}, inplace = True)
df2.rename(columns = {'CALC': 'alcFreq'}, inplace = True)
df2.rename(columns = {'MTRANS': 'transp'}, inplace = True)
df2.rename(columns = {'NObeyesdad': 'obesityLvl'}, inplace = True)

correlationMatrix = df2[['weight', 'age', 'height']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlationMatrix, annot=True, cmap ='Greens', fmt=".2f", linewidths=.5)
plt.title("Correlation Heatmap: Weight, Age, and Height")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df2['height'], y=df2['weight'], color='seagreen')
plt.title('Correlation between Height and Weight Using Scatter Plot', color='black')
plt.xlabel('Height (m)', color='black')
plt.ylabel('Weight (kg)', color='green')
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Family History of Overweight vs Obesity Level
contingencyTableFamHist = pd.crosstab(df2['famHistOverweight'], df2['obesityLvl'])
chi2, pFamHist, df, expected = chi2_contingency(contingencyTableFamHist)
print(f"p-value for Family History of Overweight vs ObesityLvl: {pFamHist}")
if pFamHist < 0.05:
    print(f"Since {pFamHist} < .05 there IS a significant association between family history of overweight and obesity level.\n")
else:
    print(f"Since {pFamHist} > .05 there IS NO significant association between family history of overweight and obesity level.\n")

from scipy.stats import chi2_contingency

# 1. Smoking Status vs Obesity Level
contingencyTableSmoking = pd.crosstab(df2['smoke'], df2['obesityLvl'])
chi2_smoking, pSmoking, df_smoking, expected_smoking = chi2_contingency(contingencyTableSmoking)
print(f"p-value for Smoking Status vs ObesityLvl: {pSmoking}")
if pSmoking < 0.05:
    print(f"Since {pSmoking} < .05 there IS a significant association between smoking status and obesity level.\n")
else:
    print(f"Since {pSmoking} > .05 there IS NO significant association between smoking status and obesity level.\n")

# 2. Alcohol Frequency vs Obesity Level
contingencyTableAlcohol = pd.crosstab(df2['alcFreq'], df2['obesityLvl'])
chi2_alcohol, pAlcohol, df_alcohol, expected_alcohol = chi2_contingency(contingencyTableAlcohol)
print(f"p-value for Alcohol Frequency vs ObesityLvl: {pAlcohol}")
if pAlcohol < 0.05:
    print(f"Since {pAlcohol} < .05 there IS a significant association between alcohol frequency and obesity level.\n")
else:
    print(f"Since {pAlcohol} > .05 there IS NO significant association between alcohol frequency and obesity level.\n")

# 3. Weight Category vs Obesity Level
contingencyTableWeight = pd.crosstab(df2['weight'], df2['obesityLvl'])
chi2_weight, pWeight, df_weight, expected_weight = chi2_contingency(contingencyTableWeight)
print(f"p-value for Weight Category vs ObesityLvl: {pWeight}")
if pWeight < 0.05:
    print(f"Since {pWeight} < .05 there IS a significant association between weight category and obesity level.\n")
else:
    print(f"Since {pWeight} > .05 there IS NO significant association between weight category and obesity level.\n")
```

**For the p values above here are discussion points**

**Family History of Overweight**
- **Can reflect genetic predisposition, dietary habits, and physical activity levels. Genetics can influence metabolism, fat distribution, appetites, etc, all of which play a role in obesity.**
- **Family history can factor in more than just genetic factors but also shared lifestyle habits. Brings up the nature vs nurture argument**

**Smoking**
- **Smoking is linked to alterations in metabolism and appetite regulation. Nicotine can lead to lifestyles with poorer diets, lower lung capacity limiting physical activity, playing a role in obesity**
- **Interesting because smoking is linked to appetite suppression**

**Alcohol Frequency**
- **Alcohol is high in calories so high consumption increases overall caloric intake. Drinking also leads to poor dietary choices and affects metabolic processes which can contribute to obesity**
- **Alcoholic intake is often underreported in surveys. Alcohol consumption can be closely tied to other behaviors like social eating, reduced physical activity, etc which can contribute to obesity making it difficult to isolate the effects of alcohol alone**

**Weight**
- **The weight category to be classified as obese is linked to the weight. So it makes sense that the heavier someone is, the more likely they are to be classified as obese**
- **Weight gain isn't always a bad thing. Many bodybuilders contain large amounts of muscle mass, classifying them in the obese category while having healthy bodies.**
