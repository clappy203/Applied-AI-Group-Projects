```python
import pandas as pd

df = pd.read_csv('obesity.csv')

df.head()
```

```python
#renamed columns
df.rename(columns = {'Gender': 'gender'}, inplace = True)
df.rename(columns = {'Age': 'age'}, inplace = True)
df.rename(columns = {'Height': 'height'}, inplace = True)
df.rename(columns = {'Weight': 'weight'}, inplace = True)
df.rename(columns = {'family_history_with_overweight': 'famHistOverweight'}, inplace = True)
df.rename(columns = {'HiCalFood': 'highCalFood'}, inplace = True)
df.rename(columns = {'FCVC': 'veggies'}, inplace = True)
df.rename(columns = {'numMealPerDay': 'numMealsPerDay'}, inplace = True)
df.rename(columns = {'CAEC': 'foodBetweenMeals'}, inplace = True)
df.rename(columns = {'SMOKE': 'smoke'}, inplace = True)
df.rename(columns = {'CH2O': 'h20PerDay'}, inplace = True)
df.rename(columns = {'SCC': 'trackCals'}, inplace = True)
df.rename(columns = {'FAF': 'physActiFreq'}, inplace = True)
df.rename(columns = {'TUE': 'techFreq'}, inplace = True)
df.rename(columns = {'CALC': 'alcFreq'}, inplace = True)
df.rename(columns = {'MTRANS': 'transp'}, inplace = True)
df.rename(columns = {'NObeyesdad': 'obesityLvl'}, inplace = True)


df.head(25)
```

```python
df['gender'] = df['gender'].replace('Female','F')
df['gender'] = df['gender'].replace('Male','M')
df.head(25)
```

```python
df['famHistOverweight'] = df['famHistOverweight'].replace('yes',1)
df['famHistOverweight'] = df['famHistOverweight'].replace('no',0)
df.head(25)
```

```python
df['highCalFood'] = df['highCalFood'].replace('yes',1)
df['highCalFood'] = df['highCalFood'].replace('no',0)
df['smoke'] = df['smoke'].replace('yes',1)
df['smoke'] = df['smoke'].replace('no',0)
df['trackCals'] = df['trackCals'].replace('yes',1)
df['trackCals'] = df['trackCals'].replace('no',0)
df.head(25)
```

```python
uniqueVals = df['obesityLvl'].unique()

print(uniqueVals)
```

```python
df['obesityLvl'] = df['obesityLvl'].replace('Normal_Weight','NormalWeight')
df['obesityLvl'] = df['obesityLvl'].replace('Insufficient_Weight','InsufficientWeight')
df['obesityLvl'] = df['obesityLvl'].replace('Overweight_Level_I','OverweightLvl1')
df['obesityLvl'] = df['obesityLvl'].replace('Overweight_Level_II','OverweightLvl2')
df['obesityLvl'] = df['obesityLvl'].replace('Obesity_Type_I','ObesityType1')
df['obesityLvl'] = df['obesityLvl'].replace('Obesity_Type_II','ObesityType2')
df['obesityLvl'] = df['obesityLvl'].replace('Obesity_Type_III','ObesityType3')

df.head(25)
```
